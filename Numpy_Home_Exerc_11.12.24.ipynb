```python
import numpy as np
```

```python
##################### Exercise 1:

# 1. Create a NumPy array of size 4x3 and filled with random numbers from 10 to
# 98 (inclusive). Calculate the sum of all elements, the sum of each row and
# each column.

array = np.random.randint(10, 99, size=(4, 3));
print("Matrix:");
print(array);
```

```python
######## EXTRA #####

import matplotlib.pyplot as plt
import seaborn as sns

array = np.random.randint(10, 99, size=(4, 3))
print("Matrix:")
print(array)

total_sum = np.sum(array)

row_sums = np.sum(array, axis=1)

col_sums = np.sum(array, axis=0)

print("\nTotal sum of all elements:", total_sum)
print("Sum of each row:", row_sums)
print("Sum of each column:", col_sums)
####################################################################

# seaborn heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(array, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.title("Heatmap of the 4x3 Matrix")
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(range(1, 5), row_sums, color='skyblue')
axes[0].set_title("Sum of Each Row")
axes[0].set_xlabel("Row Number")
axes[0].set_ylabel("Sum")

axes[1].bar(range(1, 4), col_sums, color='lightgreen')
axes[1].set_title("Sum of Each Column")
axes[1].set_xlabel("Column Number")
axes[1].set_ylabel("Sum")

plt.tight_layout()
plt.show()
```

```python
################## Exercise 2:

# 1. Create a NumPy array of size 10 filled with random numbers from 0 to 10.
# Count how many of them are even numbers.

random_numbers = np.random.randint(0, 11, size=10);

even_count = np.sum(random_numbers % 2 == 0);

print("Random numbers:", random_numbers);
print("Number of even numbers:", even_count);
```

```python
# Ex.4 - 2. Filter the array for values greater than 50 or lower than 20

filtered_array = array[(array > 50) | (array < 20)]
print("Filtered array (values > 50 or < 20):", filtered_array);
```

```python
########### EXTRA ###############

array = np.linspace(0, 100, 10);

filtered_array = array[(array > 50) | (array < 20)]

print("Filtered array (values > 50 or < 20):");
print(filtered_array);

plt.figure(figsize=(8, 5));
sns.barplot(x=np.arange(len(filtered_array)), y=filtered_array, color='lightcoral');
plt.title("Bar Plot of Filtered Array (values > 50 or < 20)");
plt.xlabel("Index");
plt.ylabel("Value");
plt.show();

plt.figure(figsize=(8, 5));
sns.lineplot(x=np.arange(len(filtered_array)), y=filtered_array, marker='o', color='darkorange');
plt.title("Line Plot of Filtered Array (values > 50 or < 20)");
plt.xlabel("Index");
plt.ylabel("Value");
plt.xticks(np.arange(len(filtered_array)));
plt.grid(True);
plt.show();
```

```python
# Ex.4 - 3. Reshape the array into a possible 2D matrix if possible

size = filtered_array.size
possible_shapes = [(rows, size // rows) for rows in range(1, size + 1) if size % rows == 0]

if possible_shapes:
    reshaped_matrix = filtered_array.reshape(possible_shapes[0])
    print(f"Reshaped array to shape {possible_shapes[0]}:\n", reshaped_matrix)
else:
    print("The array cannot be reshaped into a 2D matrix.");
```

```python
############## EXTRA #####################

array = np.linspace(0, 100, 10);

filtered_array = array[(array > 50) | (array < 20)]

size = filtered_array.size
possible_shapes = [(rows, size // rows) for rows in range(1, size + 1) if size % rows == 0]

if possible_shapes:
    reshaped_matrix = filtered_array.reshape(possible_shapes[0]);
    print(f"Reshaped array to shape {possible_shapes[0]}:\n", reshaped_matrix);

    # Visualizing the reshaped matrix using a heatmap
    plt.figure(figsize=(8, 6));
    sns.heatmap(reshaped_matrix, annot=True, fmt='g', cmap='coolwarm', cbar=True);
    plt.title(f"Heatmap of Reshaped Matrix\nShape: {possible_shapes[0]}");
    plt.show();
else:
    print("The array cannot be reshaped into a 2D matrix.");
```
